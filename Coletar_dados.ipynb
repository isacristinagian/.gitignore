Analisando índice beta e alfa para um determinado fundo:

```python
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
Fundos['Nome_fundo'] = le.fit_transform(Fundos['Nome_fundo'])

y_data = Fundos['Retorno_60_meses'].values
x_data = Fundos['Nome_fundo'].values.reshape(-1,1)

model = LinearRegression()
model.fit(x_data, y_data)

beta, alfa = np.polyfit(x=Fundos['Nome_fundo'], y=Fundos['Retorno_60_meses'], deg=1)

beta = model.coef_[0]
alfa = model.intercept_
print(beta, alfa)
```

# ***Calculando o Drawdown de um ativo***

OBS: IXIC = Nasdaq composite (índice)



```python
pip install yfinance
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import os
```

```python
asset =yf.download('^IXIC', start='2023-01-01', interval='1D')
asset.drop(['High', 'Low','Open','Volume'], axis=1, inplace=True)
asset.head()
```

```python
df = pd.DataFrame(asset)
df.head()

```

```python
df['Pico'] = df['Close'].cummax()

df.head()
```

```python
df['drawdown'] = ((df['Close'].squeeze() - df['Pico'].squeeze()) / df['Pico'].squeeze())*100
df.head()
```

```python
df['drawdown'].plot(figsize=(12,4), title='Drawdown')
```
